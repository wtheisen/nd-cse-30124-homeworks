````python
##### WRITE DIJKSTRAS_MULTI_MODE AND PRINT_BEST_PATH_TO_ESTATE FUNCTIONS #####
def dijkstra_multi_mode(nodes_df, edges_df, start):
    """
    Dijkstra's algorithm with support for multiple transport modes.

    Args:
    - nodes_df: DataFrame with columns [node_id, name].
    - edges_df: DataFrame with columns [src, dst, taxi_min, train_min, ferry_min, plane_min].
    - start: Starting node.

    Returns:
    - distances: Dictionary with the shortest distance to each node from the start.
    - paths: Dictionary with the best path to each node.
    - nodes_visited: Count of how many nodes were visited.
    - edges_evaluated: Count of how many edges were evaluated.
    """

    # Priority queue holds (distance, vertex)
    priority_queue = []
    heapq.heappush(priority_queue, (0, start))

    distances = {int(node_id): float('infinity') for node_id in nodes_df["node_id"]}
    distances[start] = 0

    paths = {int(node_id): [] for node_id in nodes_df["node_id"]}
    paths[start] = [start]

    visited = set()
    nodes_visited = 0
    edges_evaluated = 0

    while priority_queue:
        current_distance, current_vertex = heapq.heappop(priority_queue)

        if current_vertex in visited:
            continue

        visited.add(current_vertex)
        nodes_visited += 1

        neighbors = edges_df[edges_df["src"] == current_vertex]
        for edge in neighbors.itertuples(index=False):
            edges_evaluated += 1
            costs = [edge.taxi_min, edge.train_min, edge.ferry_min, edge.plane_min]
            valid_costs = [cost for cost in costs if cost > 0]
            if not valid_costs:
                continue
            min_cost = min(valid_costs)

            distance = current_distance + min_cost
            neighbor = int(edge.dst)

            if distance < distances[neighbor]:
                distances[neighbor] = distance
                paths[neighbor] = paths[current_vertex] + [neighbor]
                heapq.heappush(priority_queue, (distance, neighbor))

    return distances, paths, nodes_visited, edges_evaluated


def print_best_path_to_estate(paths, edges_df, nodes_lookup, start, destination):
    """
    Prints the best path from the start node to the destination node with transport modes and times,
    without repeating node names.

    Args:
    - paths: Dictionary of shortest paths to each node.
    - edges_df: DataFrame with columns [src, dst, taxi_min, train_min, ferry_min, plane_min].
    - nodes_lookup: Series mapping node_id -> name.
    - start: The start node number.
    - destination: The destination node number.
    """
    mode_names = ["TAXI", "TRAIN", "FERRY", "FLIGHT"]

    path = paths.get(destination, None)
    if not path or path[0] != start:
        print("No valid path found.")
        return

    path_segments = [nodes_lookup.loc[start]]
    total_time = 0

    for i in range(len(path) - 1):
        current = path[i]
        next_node = path[i + 1]
        edge = edges_df[(edges_df["src"] == current) & (edges_df["dst"] == next_node)].iloc[0]
        costs = [edge.taxi_min, edge.train_min, edge.ferry_min, edge.plane_min]
        valid_costs = [cost for cost in costs if cost > 0]
        if not valid_costs:
            continue
        min_cost = min(valid_costs)
        mode_index = costs.index(min_cost)
        mode = mode_names[mode_index]
        time = costs[mode_index]

        path_segments.append(f" ---[{mode}, {time} min]--> {nodes_lookup.loc[next_node]}")
        total_time += time

    print(f"Best path from {nodes_lookup.loc[start]} to {nodes_lookup.loc[destination]}:")
    print("".join(path_segments))
    print(f"Total travel time: {total_time} minutes")
````

````python
##### RUN FUNCTIONS TO FIND PATH #####
# Using the previously defined `nodes_df` and `edges_df`
start_node = 0     # Central Park

# Find the shortest path
distances, paths, nodes_visited, edges_evaluated = dijkstra_multi_mode(nodes_df, edges_df, start_node)
print(f"Number of nodes visited: {nodes_visited}")
print(f"Number of edges evaluated: {edges_evaluated}")

# Print the best path
print_best_path_to_estate(paths, edges_df, nodes_lookup, start=0, destination=13)
````

### *Task Description*

Write an A* implementation to make use of the heuristic information provided in the table below. Calculate:

   * The number of nodes visited
   * The number of edges evaluated

### *Code Implemetation*

````python
import heapq
import os
import pandas as pd

heuristics_df = pd.read_csv(os.path.join(DATA_DIR, "heuristics.csv"))
heuristics_df["node_id"] = heuristics_df["node_id"].astype(int)
heuristics_lookup = heuristics_df.set_index("node_id")["heuristic"]

def a_star(nodes_df, edges_df, heuristics_lookup, start, destination):
    """
    A* algorithm for finding the shortest path in a graph.

    Args:
    - nodes_df: DataFrame with columns [node_id, name].
    - edges_df: DataFrame with columns [src, dst, taxi_min, train_min, ferry_min, plane_min].
    - heuristics_lookup: Series mapping node_id -> heuristic.
    - start: The starting node.
    - destination: The destination node.

    Returns:
    - path: List of nodes representing the shortest path from start to destination.
    - cost: Total travel cost of the shortest path.
    - nodes_visited: Number of nodes visited.
    - edges_evaluated: Number of edges evaluated.
    """
    priority_queue = []
    heapq.heappush(priority_queue, (heuristics_lookup.loc[start], start, 0, [start]))

    g_costs = {int(node_id): float('inf') for node_id in nodes_df["node_id"]}
    g_costs[start] = 0

    visited = set()
    nodes_visited = 0
    edges_evaluated = 0

    while priority_queue:
        _, current_node, current_g_cost, path = heapq.heappop(priority_queue)

        if current_node in visited:
            continue

        visited.add(current_node)
        nodes_visited += 1

        if current_node == destination:
            return path, current_g_cost, nodes_visited, edges_evaluated

        neighbors = edges_df[edges_df["src"] == current_node]
        for edge in neighbors.itertuples(index=False):
            edges_evaluated += 1
            costs = [edge.taxi_min, edge.train_min, edge.ferry_min, edge.plane_min]
            valid_costs = [cost for cost in costs if cost > 0]
            if not valid_costs:
                continue
            transport_cost = min(valid_costs)
            neighbor = int(edge.dst)

            new_g_cost = current_g_cost + transport_cost
            if new_g_cost < g_costs[neighbor]:
                g_costs[neighbor] = new_g_cost
                f_cost = new_g_cost + heuristics_lookup.loc[neighbor]
                heapq.heappush(priority_queue, (f_cost, neighbor, new_g_cost, path + [neighbor]))

    return None, float('inf'), nodes_visited, edges_evaluated

# Find the best path from Central Park (0) to William Theisen-Floyd Estate (13)
path, cost, nodes_visited, edges_evaluated = a_star(nodes_df, edges_df, heuristics_lookup, start=0, destination=13)

print(f"Number of nodes visited: {nodes_visited}")
print(f"Number of edges evaluated: {edges_evaluated}")

print_best_path_to_estate({13: path}, edges_df, nodes_lookup, start=0, destination=13)
````

### *Short Answer Questions (2 pts.)*

#### Were the results of the two algorithms the same or different? Why do you think this is? (2 pt.)



*  The two algorithims produced the same final path, but A* did so by visiting less nodes and evaluating less edges. This is because Dijkstraâ€™s algorithm expands nodes purely in order of increasing path cost g(n), exploring until all shortest paths are guaranteed. In contrast, A* uses both the path cost g(n) and a heuristic estimate h(n) of the remaining distance. This heuristic guides the search toward the destination, reducing the amount of exploration.
*   Since the heuristic was admissible (it never overestimated the true cost to the goal state), A* guaranteed the same optimal path as Dijkstra but with better efficiency.

In the code cell below implement the backtracking algorithm for constraint satisfaction problems to solve the murder mystery!

### *Code Implemetation*

````python
import os
import pandas as pd

people_df = pd.read_csv(os.path.join(DATA_DIR, "people.csv"))
rooms_df = pd.read_csv(os.path.join(DATA_DIR, "rooms.csv"))
weapons_df = pd.read_csv(os.path.join(DATA_DIR, "weapons.csv"))

people = people_df["person"].tolist()
rooms = rooms_df["room"].tolist()
weapons = weapons_df["weapon"].tolist()

# Set of all assignments
assignments = []
num_checks = 0

# Define domains
domains = {
    person: set(rooms) for person in people
}
domains.update({
    room: set(weapons) for room in rooms
})

# Clues as constraints
def is_valid(assignment):
    """
    Check if the current assignment satisfies all clues.
    """
    global num_checks
    num_checks += 1

    # Extract assignments for people, rooms, and weapons
    person_to_room = {person: room for person, room, weapon in assignment}
    person_to_weapon = {person: weapon for person, room, weapon in assignment}
    room_to_weapon = {room: weapon for person, room, weapon in assignment}

    # TODO: Clue 1: The person in the kitchen was not found with the rope, knife, or firearm.
    if room_to_weapon.get("Kitchen", "") in ["Rope", "Knife", "Firearm"]:
        return False

    # TODO: Clue 2: Miss Scarlet was either in the study or the bathroom; Professor Plum was in the other.
    if person_to_room.get("Miss Scarlet", "") not in ["Study", "Bathroom", ""]:
        return False
    if person_to_room.get("Professor Plum", "") not in ["Study", "Bathroom", ""]:
        return False

    # TODO: Clue 3: The person with the bag, who was not Miss Scarlet nor Professor Plum, was not in the bathroom nor the dining room.
    for person, room, weapon in assignment:
        if weapon == "Bag" and (person == "Miss Scarlet" or person == "Professor Plum"):
            return False
        if weapon == "Bag" and room in ["Bathroom", "Dining Room"]:
            return False

    # TODO: Clue 4: The person with the rope was found in the study.
    for person, room, weapon in assignment:
        if weapon == "Rope" and room != "Study":
            return False

    # TODO: Clue 5: The weapon in the living room was found with either Miss Scarlet or Mrs. White.
    living_room_weapon = room_to_weapon.get("Living Room", "")
    if living_room_weapon and person_to_room.get("Miss Scarlet", "") != "Living Room" and         person_to_room.get("Mrs. White", "") != "Living Room":
        return False

    # TODO: Clue 6: The knife was not in the dining room.
    if room_to_weapon.get("Dining Room", "") == "Knife":
        return False

    # TODO: Clue 7: Mrs. Peacock was not in the study nor the pantry.
    if person_to_room.get("Mrs. Peacock", "") in ["Study", "Pantry"]:
        return False

    # TODO: Clue 8: The firearm was in the room with Mrs. White.
    if person_to_weapon.get("Mrs. White", "") not in ["Firearm", ""]:
        return False

    # TODO: Final clue: Mr. Theisen-Floyd was gassed in the pantry, and the suspect found in that room is the murderer.
    if room_to_weapon.get("Pantry", "") not in ["Gas", ""]:
        return False

    return True

# Backtracking algorithm
def backtrack(assignment):
    """
    Perform backtracking to solve the puzzle.
    """
    # TODO: Check length of assignment
    if len(assignment) == 6:
        return assignment

    # TODO: Try all combinations of people, rooms, and weapons
    for person in people:
        # Skip of person not already assigned
        if person not in [p for p, r, w in assignment]:
            for room in rooms:
                # skip if rooms already assigned
                if room not in [r for p, r, w in assignment]:
                    for weapon in weapons:
                        # Skip of weapon already assigned
                        if weapon not in [w for p, r, w in assignment]:
                            # Create new assignment
                            new_assignment = assignment + [(person, room, weapon)]
                            # Only recurse if this partial assignment is still valid
                            if is_valid(new_assignment):
                                result = backtrack(new_assignment)
                                if result:
                                    return result

    # TODO: If no combination was found that works, return none
    return None

# Find the solution (Note: my solution runs almost instantly so if yours doesn't you have a bug)
solution = backtrack(assignments)

# Print the solution
if solution:
    print("Solution found:")
    for person, room, weapon in solution:
        print(f"{person} was in the {room} with the {weapon}")
    print('Num Checks', num_checks)
else:
    print("No solution found.")
````
